```python
#step 2
ords_prods_merge.groupby('department_id').agg({'order_number': 'mean', 'order_dow': 'mean', 'order_hour_of_day': 'mean', 'prices': 'mean'})
```

- order times are very similar across each departement
- department 16 is by far the most expensive, followed by 12

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)  
```

```python
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'high_spender'
```

```python
#step 7
ords_prods_merge['median_days_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] > 20, 'order_frequency'] = 'non_frequent_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_order'] > 10) & 
                     (ords_prods_merge['median_days_since_order'] <= 20), 'order_frequency'] = 'regular_customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] <= 10, 'order_frequency'] = 'frequent_customer'
```
